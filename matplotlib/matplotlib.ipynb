```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line Graph

### Titles, labels, font, legend, color, lable

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, 
         label='2x', 
         c='y', linewidth=2, # color='yellow'
         marker='.', 
         ms=10,              # markersize=10
         mec='r',            # markeredgecolor='red'
         mew='1',            # markeredgewidth='1'
         mfc='b',            # markerfacecolor='b;ie'
        )

plt.title("Our first graph", fontdict={'fontname': 'Ani', 'fontsize': 25})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,2.5,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show() # Removes [<matplotlib.lines.Line2D at 0x7ff44cede950>] from display
```

### Multiple functions, create y with formula

#### Even shorter hand

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize=(5,3), dpi=150)

plt.plot(x,y, 'b^--', label='2x')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, 'r',label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.title("Our first graph", fontdict={'fontname': 'Ani', 'fontsize': 25})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,2.5,3,4])

plt.legend()

plt.savefig('mygraph.png', dpi=300)

plt.show() # Removes [<matplotlib.lines.Line2D at 0x7ff44cede950>] from display
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

patterns = ['/','O','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

## Real World Examples

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,7))
plt.title("Gas Prices", fontdict={'fontweight': 'bold','fontsize': '15'})

plt.plot(gas.Year, gas.USA, 'r.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'b.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

# countries = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#   if country in countries:
#     plt.plot(gas.Year, gas[country], marker='.', label=country)
    
# Unlike the video I had to specify the label

plt.xticks(gas.Year[::3].tolist()+[2011])
# plt.xticks(gas.Year[::3]+[3])  # This also works


plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()

plt.savefig(f"Gas Prices", dpi=250)

plt.show()
```

### Histograms

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#83c142')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in Fifa 2018', fontdict={'fontsize': 15})
plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abc123', '#aabbcc']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f%%')

plt.title('Foot preference of FIFA players')

plt.show()
```

```python
fifa.Weight = [float(str(x).strip('lbs')) for x in fifa.Weight]
# Video used int() which failed for me
fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]

plt.style.use('ggplot')

labels = ['Under 125', '125-149', '150-174', '175-199', '200+']
weights = [light, light_medium, medium, medium_heavy, heavy]
explode = (.4,.2,0,0,.4)

plt.title("Weight Distribution in lbs")

plt.pie(weights, 
        labels=labels, 
        autopct='%.2f %%', 
        pctdistance=.8,
        explode=explode
       )

plt.show()
```

### Box & Whiskers Chart

```python
fifa.head(8)
```

```python
plt.style.use('default')

plt.figure(figsize=(6,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], 
                    labels=labels, 
                    patch_artist=True, 
                    medianprops={'linewidth':2}
                   )
# patch_artist = True required for facecolor

for box in boxes['boxes']:
  box.set(color='#4286f4', 
          linewidth=2,
          facecolor='#e0e0e0',
         )

plt.title('Professional Soccer Team Rating')
plt.ylabel('FIFA Overall Rating')



plt.show()
```
